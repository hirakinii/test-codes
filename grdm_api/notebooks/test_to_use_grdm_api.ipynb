# load a node with title

例えば file もタイトルで取ってこれる？

```python
endpoint: str = f"{GRDM_URL}/nodes/?filter[category]=project&filter[title]=dmp"
headers = {
    'Accept': 'application/vnd.api+json',
    "Authorization": f"Bearer {token}",
    'Content-Type': "application/json",
}

node_ids: list = []
node_title: list = []
next_url: str = endpoint
while next_url:
    response = requests.get(
        endpoint, headers=headers,
        timeout=30
    )
    response.raise_for_status()
    node_ids.extend([node["id"] for node in response.json()["data"]])
    node_title.extend([node["attributes"]["title"] for node in response.json()["data"]])
    print("current ids:", len(node_ids))
    next_url = response.json().get("links", {}).get("next", "")
    # time.sleep(0.25)
```
